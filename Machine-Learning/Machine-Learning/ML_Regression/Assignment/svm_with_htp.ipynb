```python
import pandas as pd
```

```python
df=pd.read_csv("insurance_pre.csv")
df
```

```python
df=pd.get_dummies(df,dtype=int,drop_first=True)
df
```

```python
df.columns
```

```python
independent=df[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=df[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.20,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
X_train
```

```python
X_test
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='sigmoid',C=1)
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='rbf',C=1)
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='poly',C=1)
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='linear',C=1)
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
