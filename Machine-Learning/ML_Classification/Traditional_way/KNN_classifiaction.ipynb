```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset = dataset.drop(columns=['User ID'],axis=1)
```

```python
dataset
```

```python
dataset.columns
```

```python
X=dataset[['Age', 'EstimatedSalary','Gender_Male']]
y=dataset[['Purchased']]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=1/3,random_state=0)
k_neighbour=KNeighborsClassifier(n_neighbors=5, weights='uniform', algorithm='auto', leaf_size=30, p=2, metric='minkowski', metric_params=None, n_jobs=None)
k_neighbour.fit(X_train,y_train)
```

```python
y_pred=k_neighbour.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
class_report=classification_report(y_test,y_pred)
print(class_report)
```
